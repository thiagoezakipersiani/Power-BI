```python
import fundamentus
df = fundamentus.get_resultado()
print(df)
```

```python
#Trazendo somente PL maiores que 0
df_pl = df[(df.pl > 0)]

#Ordenando pelo pl
df_pl=df_pl.sort_values(by="pl", ascending=True)

#Criando um array com o numero da linha para ser inserido na coluna ranking
i=1
coluna_ranking =  []
for n in range(len(df_pl.index)):
    coluna_ranking.append(i)
    i+=1

df_pl.insert(loc=20, column='ranking', value=coluna_ranking)

print(df_pl)
```

```python
#Trazendo somente dy maiores que 0
df_dy = df[(df.dy > 0)]

#Ordenando pelo dy
df_dy=df_dy.sort_values(by="dy", ascending=False)

#Criando um array com o numero da linha para ser inserido na coluna ranking
i=1
coluna_ranking =  []
for n in range(len(df_dy.index)):
    coluna_ranking.append(i)
    i+=1

df_dy.insert(loc=20, column='ranking', value=coluna_ranking)

print(df_dy)
```
