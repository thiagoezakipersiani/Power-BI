```python
import fundamentus
import pandas as pd

data = fundamentus.get_resultado()
print(data)

df = data[(data.liq2m > 0) & (data.patrliq > 1500000) ]

df = pd.DataFrame(df, columns=['cotacao', 'pl', 'pvp', 'psr', 'dy', 'pa', 'pcg', 'pebit', 'pacl',
         'evebit', 'evebitda', 'mrgebit', 'mrgliq', 'roic', 'roe', 'liqc',
         'liq2m', 'patrliq', 'divbpatr', 'c5y'])

df

#Trazendo somente PL maiores que 0
df_pl = df[(df.pl > 0)]

#Ordenando pelo pl
df_pl=df_pl.sort_values(by="pl", ascending=True)

#Criando um array com o numero da linha para ser inserido na coluna ranking
i=1
coluna_ranking =  []
for n in range(len(df_pl.index)):
    coluna_ranking.append(i)
    i+=1

df_pl.insert(loc=df_pl.shape[1], column='ranking_pl', value=coluna_ranking)

df_pl = (df_pl['ranking_pl'])

#---------------------------------------------

#Trazendo somente dy maiores que 0
df_dy = df[(df.dy > 0)]

#Ordenando pelo dy
df_dy=df_dy.sort_values(by="dy", ascending=False)

#Criando um array com o numero da linha para ser inserido na coluna ranking
i=1
coluna_ranking =  []
for n in range(len(df_dy.index)):
    coluna_ranking.append(i)
    i+=1

df_dy.insert(loc=df_dy.shape[1], column='ranking_dy', value=coluna_ranking)

df_dy = (df_dy['ranking_dy'])

#---------------------------------------------

#Trazendo informações do ROIC
df_roic = df[(df.roic > 0)]

df_roic = df_roic.sort_values(by='roic',ascending=False)

coluna_ranking = []
i=1

for n in range(len(df_roic.index)):
    coluna_ranking.append(i)
    i+=1
    
df_roic.insert(loc=df_roic.shape[1],column='ranking_roic', value=coluna_ranking)

df_roic = (df_roic['ranking_roic'])

#---------------------------------------------

#Trazendo informações do PSR
df_psr = df[(df.psr > 0) & (df.psr < 1)]

df_psr = df_psr.sort_values(by='psr',ascending=True)

i=1
coluna_ranking = []
for n in range(len(df_psr.index)):
    coluna_ranking.append(i)
    i+=1
    
df_psr.insert(loc=df_psr.shape[1], column='ranking_psr', value=coluna_ranking)

df_psr = (df_psr['ranking_psr'])


#---------------------------------------------

#Trazendo informações do ROE
df_roe = df[(df.roe > 0)]

df_roe = df_roe.sort_values(by='roe',ascending=False)

coluna_ranking = []
i=1

for n in range(len(df_roe.index)):
    coluna_ranking.append(i)
    i+=1
    
df_roe.insert(loc=df_roe.shape[1],column='ranking_roe', value=coluna_ranking)

df_roe = (df_roe['ranking_roe'])

#Concatenando todos os dataframes com a coluna ranking de cada
df_final = pd.concat([df, df_pl, df_roe,df_psr,df_dy], axis=1)

#Criando a coluna de ranking final
df_final['soma_ranking'] = df_final['ranking_pl'] + df_final['ranking_dy'] + df_final['ranking_psr'] + df_final['ranking_roe']

print(df_final)

df_final = df_final[df_final.soma_ranking > 0]

df_final = df_final.sort_values(by='soma_ranking',ascending=True)

coluna_ranking = []

i=1

for n in range(len(df_final.index)):
    coluna_ranking.append(i)
    i+=1

df_final.insert(loc=df_final.shape[1],column='ranking_final', value=coluna_ranking)

print(df_final)

df_final.to_csv("./Excel/ranking_final.csv", sep=";")
```
