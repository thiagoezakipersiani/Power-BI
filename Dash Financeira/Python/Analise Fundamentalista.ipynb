```python
import fundamentus
df = fundamentus.get_resultado()
print(df)

df = df[(df.liq2m > 0) & (df.patrliq > 1500000) ]

df.to_csv('./Excel/resultados.csv', sep=';')

#---------------------------------------------

#Trazendo somente PL maiores que 0
df_pl = df[(df.pl > 0)]

#Ordenando pelo pl
df_pl=df_pl.sort_values(by="pl", ascending=True)

#Criando um array com o numero da linha para ser inserido na coluna ranking
i=1
coluna_ranking =  []
for n in range(len(df_pl.index)):
    coluna_ranking.append(i)
    i+=1

df_pl.insert(loc=20, column='ranking_pl', value=coluna_ranking)

print(df_pl['ranking_pl'])

df_pl['ranking_pl'].to_csv("./Excel/resultados_pl.csv", sep=';')

#---------------------------------------------

#Trazendo somente dy maiores que 0
df_dy = df[(df.dy > 0)]

#Ordenando pelo dy
df_dy=df_dy.sort_values(by="dy", ascending=False)

#Criando um array com o numero da linha para ser inserido na coluna ranking
i=1
coluna_ranking =  []
for n in range(len(df_dy.index)):
    coluna_ranking.append(i)
    i+=1

df_dy.insert(loc=20, column='ranking_dy', value=coluna_ranking)

print(df_dy['ranking_dy'])

df_dy['ranking_dy'].to_csv('./Excel/resultados_dy.csv', sep=';')

#---------------------------------------------

#Trazendo informações do ROIC
df_roic = df[(df.roic > 0)]

df_roic = df_roic.sort_values(by='roic',ascending=False)

coluna_ranking = []
i=1

for n in range(len(df_roic.index)):
    coluna_ranking.append(i)
    i+=1
    
df_roic.insert(loc=20,column='ranking_roic', value=coluna_ranking)

print(df_roic['ranking_roic'])

df_roic['ranking_roic'].to_csv('./Excel/resultados_roic.csv',sep=';')

#---------------------------------------------

#Trazendo informações do PSR
df_psr = df[(df.psr > 0) & (df.psr < 1)]

df_psr = df_psr.sort_values(by='psr',ascending=True)

i=1
coluna_ranking = []
for n in range(len(df_psr.index)):
    coluna_ranking.append(i)
    i+=1
    
df_psr.insert(loc=20, column='ranking_psr', value=coluna_ranking)

print(df_psr['ranking_psr'])

df_psr['ranking_psr'].to_csv('./Excel/resultados_psr.csv',sep=';')

#---------------------------------------------

#Trazendo informações do ROE
df_roe = df[(df.roe > 0)]

df_roe = df_roe.sort_values(by='roe',ascending=False)

coluna_ranking = []
i=1

for n in range(len(df_roe.index)):
    coluna_ranking.append(i)
    i+=1
    
df_roe.insert(loc=20,column='ranking_roe', value=coluna_ranking)

print(df_roe['ranking_roe'])

df_roe['ranking_roe'].to_csv('./Excel/resultados_roe.csv',sep=';')
```
